# Exploratory Data Analysis of the Data

A small notebook to visualize the data and how much of it there is.

Currently:

- A song of ice and fire books


```python
import os
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style="whitegrid")
```

```python
asoif_path = "../books/asoif/"
books = [f for f in os.listdir(asoif_path) if f.endswith(".txt")]

df_builder = []

for book in books:
    with open(asoif_path + book) as f:
        content = f.read()
        words = content.split()
        df_builder.append({"Book": book.split(".")[0], "Word count": len(words)})

df = pd.DataFrame(df_builder)
```

```python
plt.figure()
sns.barplot(x="Book", y="Word count", data=df)
plt.xticks(rotation=45)

print(f"In total there are {df['Word count'].sum()} words in the books")
plt.show()
```
