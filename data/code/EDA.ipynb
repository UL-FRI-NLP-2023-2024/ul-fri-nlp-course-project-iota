# Exploratory Data Analysis of the Data

A small notebook to visualize the data and how much of it there is.

Currently:

- A song of ice and fire books
  - The books have been filtered to only include the chapters: No Copyright info, ackowledgements,...
  - Harry Potter books https://www.kaggle.com/datasets/moxxis/harry-potter-lstm?resource=download


```python
import os
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style="whitegrid")
```

```python
def word_count(series, folder):
    path = os.path.join("..", folder, series)
    books = [f for f in os.listdir(path) if f.endswith(".txt")]

    df_builder = []

    for book in books:
        with open(os.path.join(path, book)) as f:
            content = f.read()
            words = content.split()
            df_builder.append({"Book": book.split(".")[0], "Word count": len(words)})

    df = pd.DataFrame(df_builder)
    df = df.sort_values(by="Book")

    plt.figure()
    sns.barplot(x="Book", y="Word count", data=df)
    plt.xticks(rotation=45)

    print(f"In total there are {df['Word count'].sum()} words in the books")
    plt.show()
```

```python
word_count("asoif", "books")
```

```python
word_count("hp", "books")
```

```python
word_count("asoif", "summaries")
word_count("hp", "summaries")
```
